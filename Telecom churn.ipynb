#### Literature Review & Hypothesis

#### The Commitment Gap
- ***The Literature:*** Research shows that "Month-to-Month" plans act like an open exit. Without a long-term contract or a cancellation fee, customers are much more likely to "shop around" and leave the moment a competitor offers a better deal.
- ***The Hypothesis:*** Customers on Month-to-Month contracts will have a significantly higher churn rate than those on one- or two-year plans.
- ***Source:*** PubMed Central (PMC) via the National Institutes of Health.

#### The "Digital Glue" Theory
- ***The Literature:*** Experts found that "Value-Added Services" (like Online Security and Tech Support) make a company "sticky." When a customer relies on you for their safety and troubleshooting, it becomes a major headache for them to switch everything over to a new provider.
- ***The Hypothesis:*** Having Online Security and Tech Support will be the strongest predictors of a customer staying, regardless of how much they pay monthly
- ***Source:*** Frontiers in Artificial Intelligence

#### The Newbie Risk Window
- ***The Literature:*** Statistics show that the first year of a service is the "danger zone." If a customer survives the initial setup and the first few bills without getting frustrated, they usually fall into a habit and stay loyal for a long time.
- ***The Hypothesis:*** Churn will be highest for customers with a Tenure of less than 12 months and will drop drastically for anyone who has been around for over two years.
- ***Source:***  Algorithms Journal (MDPI)

#### The Tech Frustration Factor
- ***The Literature:*** High-speed Fiber Optic internet is great, but if it's too complex and there's no support, it actually drives people away.
- ***The Hypothesis:*** Senior Citizens will show high churn rates if they have Fiber Optic internet but don't have Tech Support included in their package.
- ***Source:*** Frontiers in Artificial Intelligence

### Summary of Proven Hypotheses

| Hypothesis | Theory | Result | Key Evidence |
| :--- | :--- | :--- | :--- |
| **The Commitment Gap** | Month-to-Month plans act as an open exit for customers. | **Proven** | Churn rate is significantly higher (~43%) compared to 1 or 2-year contracts. |
| **The Digital Glue** | Value-added services like Tech Support make the service "sticky." | **Proven** | Customers with Online Security and Tech Support show much lower attrition. |
| **The Newbie Risk** | The first 12 months are the "danger zone" for new sign-ups. | **Proven** | Histogram shows a massive spike in churn density within the first year. |
| **Tech Frustration** | High-speed Fiber Optic users churn more without proper support. | **Proven** | Fiber Optic churn is high, but drops when bundled with Technical Support. |

### Actionable Business Recommendations

- Targeted Retention: Focus marketing efforts on customers in their first 6–12 months. Offer a "1-year anniversary" discount to push them past the high-risk tenure window.

- Bundle Strategy: Incentivize Fiber Optic users to add "Online Security" or "Tech Support" to their plan. Frame it as a "Peace of Mind" bundle to reduce tech-related frustration.

- Contract Migration: Convert Month-to-Month users to longer-term contracts by offering a small monthly discount.

- Payment Friction: Since "Electronic Check" users churned more, encourage "Auto-pay" via Credit Card or Bank Transfer to reduce payment-related service interruptions.

### Final Conclusion & Strategic Roadmap

#### By moving from a reactive to a proactive retention strategy—focusing on the first 12 months and the 'Digital Glue' of support services—the company can significantly reduce revenue leakage and increase customer lifetime value (CLV).
